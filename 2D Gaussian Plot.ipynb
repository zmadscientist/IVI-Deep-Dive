```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Generate fake data
x = np.random.normal(size=1000)
y = x * 3 + np.random.normal(size=1000)

# Calculate the point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

fig, ax = plt.subplots()
ax.scatter(x, y, c='k', s=8, alpha = .1)
ax.set_xlim(-4, 4)
ax.set_ylim(-10, 10)
ax.arrow(-2.5, -8, 5, 16, head_width=.3, head_length=.5, fc='b', ec='b')
ax.arrow(0, 0, -.3, 1.3, head_width=0.3, head_length=0.5, fc='r', ec='r')


plt.show()
```
